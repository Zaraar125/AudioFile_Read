```
import librosa
from librosa import display
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```
# we need to read the sound file
#first we need to give the path of the file
# then we have to import the library librosa
# then read the file using librosa.load
# it will return the lenght of the file and sample rate or frequency
# it will automatically do feature selection and keep good features and leave the bad ones
```

```
sound_file_path="/content/SM7_F20_H092.wav"
len_of_file,sample_rate=librosa.load(sound_file_path) 
len(len_of_file)
```

```
# we an also use the sound file library to load the sound file
```

```
len_of_file,sample_rate=sf.read(sound_file_path)
len(len_of_file)
```

```
# then after loading we have to do feature extraction
# we can use the chromagram of librosa to do feature extraction
# librosa.feature,chroma_stft    short-term-fourier-transform
# this will further clean the audio file and keep the best deatures and leave the rest
# hop_length=?
```

```
hop_length=512
chromagram=librosa.feature.chroma_stft(len_of_file,sr=sample_rate,hop_length=hop_length)
len(chromagram)
```

```
# now we have extracted the most best features
# ACCORDING TO THE CHROMA FEATURE EXTRACTION
# now using matplotlib we will plot these features
# to display we use librosa.display.specshow
# we need to explicitly import the display from librosa
```

```
plt.figure(figsize=(15,5))
display.specshow(chromagram,x_axis="time",y_axis="chroma",cmap="coolwarm")
```

```
# now we apply short term fourier transform
# this is the second type of featuer extraction
```

```
stft=np.abs(librosa.stft(len_of_file))
librosa.display.specshow(stft,sr=sample_rate,x_axis="time")
```

```
# the third type of feature extraction is the MFCC
```

```
MFCC_features=librosa.feature.mfcc(len_of_file,sr=sample_rate)
len(MFCC_features)
```

```
plt.figure(figsize=(10,5))
librosa.display.specshow(MFCC_features,x_axis="time")
```
